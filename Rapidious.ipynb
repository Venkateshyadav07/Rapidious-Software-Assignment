```python


# Data Preparation and Train-Test Split
X = df[['calories', 'protein', 'fat', 'sodium']]  # Features
y = df['dessert']  # Target

# Split the data into training and test sets
xtrain, xtest, ytrain, ytest = train_test_split(X, y, random_state=1)

# Logistic Regression Model Training
reg = LogisticRegression()
reg.fit(xtrain, ytrain)

# Predict the target values for the test set
y_pred_lr = reg.predict(xtest)

# Calculate accuracy scores
log_train = round(reg.score(xtrain, ytrain) * 100, 2)
log_accuracy_MinMax = round(accuracy_score(y_pred_lr, ytest) * 100, 2)

# Output the results
print("Training Accuracy    :", log_train, "%")
print("Model Accuracy Score :", log_accuracy_MinMax, "%")

```

```python
# Confusion Matrix Plot
cm = confusion_matrix(ytest, y_pred_lr) # Create the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=reg.classes_) # Pass the confusion matrix to the ConfusionMatrixDisplay object
disp.plot() # Display the confusion matrix
plt.title('Confusion Matrix')
plt.show()


# Feature Engineering
df["prot_on_100cal"] = round(df["protein"] / (df["calories"] / 100), 2)
print(df.head())  # Display the first few rows

# Visualization of the New Feature Distribution
a = sns.FacetGrid(df, hue="dessert", aspect=4)
a.map(sns.kdeplot, "prot_on_100cal", shade=True)
# Calculate the maximum value of 'prot_on_100cal' while excluding 'inf' values
max_value = np.nanmax(df['prot_on_100cal'].replace([np.inf, -np.inf], np.nan).dropna())
a.set(xlim=(0, max_value)) # Set the x-axis limits using the calculated maximum value
a.add_legend()
plt.show()
```

```python
# Import Required Libraries
from flask import Flask, request, jsonify
import pickle
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd

# Assume df is already loaded and prepared
# Check the column names in your DataFrame
print(df.columns)

# Calculate the 'prot_per_100cal' column
df['prot_per_100cal'] = df['protein'] / (df['calories'] / 100)

# Data Preparation (X and y variables)
X = df[["prot_per_100cal"]]  # Now using the calculated 'prot_per_100cal'
y = df['dessert']  # Target: Dessert or not

# Remove rows with NaN or infinity in 'prot_per_100cal'
X = X.replace([np.inf, -np.inf], np.nan).dropna()
y = y.loc[X.index]  # Ensure the target variable is aligned with X after NaN removal

# Split the data into training and test sets
xtrain, xtest, ytrain, ytest = train_test_split(X, y, random_state=1)

# Initialize and train the RandomForestClassifier
rforest = RandomForestClassifier(n_estimators=100)
rforest.fit(xtrain, ytrain)

# Save the trained model to a file using pickle
with open('rforest_model.pkl', 'wb') as f:
    pickle.dump(rforest, f)

# Flask application
app = Flask(__name__)

# Load the model
with open('rforest_model.pkl', 'rb') as f:
    model = pickle.load(f)

# Define route for predictions
@app.route('/predict', methods=['POST'])
def predict():
    # Get data from the request
    data = request.json
    prot_per_100cal = data.get('prot_per_100cal')  # Get input from request JSON

    # Convert input to a DataFrame
    input_data = pd.DataFrame([[prot_per_100cal]], columns=["prot_per_100cal"])

    # Predict using the loaded model
    prediction = model.predict(input_data)

    # Prepare the response
    result = {'prediction': int(prediction[0])}
    return jsonify(result)

# Run the Flask app
if __name__ == '__main__':
    app.run(debug=True)
```

```python
# Import Required Libraries
import streamlit as st
import pickle
import pandas as pd
import numpy as np
```

```python


# Load the trained model
with open('rforest_model.pkl', 'rb') as f:
    model = pickle.load(f)

# Function for prediction
def predict_dessert(prot_per_100cal):
    input_data = pd.DataFrame([[prot_per_100cal]], columns=["prot_per_100cal"])
    prediction = model.predict(input_data)
    return int(prediction[0])

# Streamlit app
def main():
    st.title("Dessert Prediction Based on Protein per 100 Calories")

    # Input from user
    prot_per_100cal = st.number_input("Enter protein per 100 calories:", min_value=0.0, step=0.1)

    # When the user clicks the "Predict" button
    if st.button("Predict"):
        # Perform prediction
        result = predict_dessert(prot_per_100cal)

        # Display result
        if result == 1:
            st.success("The item is likely a dessert!")
        else:
            st.success("The item is not a dessert.")

if __name__ == '__main__':
    main()
```
